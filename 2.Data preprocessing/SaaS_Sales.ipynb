```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
```

```python
 # Load dataset
dataset = pd.read_csv(r"D:\python\Machine Learning and Data Science Capstone Project\1.Data collection\SaaS-Sales.csv")
    
    
```

```python
dataset
```

```python
# Drop unnecessary columns
drop_cols = [
    'Row ID',
    'Order ID',
    'Contact Name',
    'Customer ID',
    'License',
    'Date Key'
]

dataset = dataset.drop(columns=drop_cols)
```

```python
# Convert Order Date to datetime
dataset['Order Date'] = pd.to_datetime(dataset['Order Date'])
    
```

```python
# =========================
# Separate numeric columns
# =========================
num_cols = ['Sales', 'Quantity', 'Discount', 'Profit']
    
```

```python
# Fill numeric missing values
num_imputer = SimpleImputer(strategy='median')

dataset[num_cols] = num_imputer.fit_transform(dataset[num_cols])
```

```python
# =========================
# Separate categorical columns
# =========================
cat_cols = dataset.select_dtypes(include=['object', 'string']).columns

# Fill categorical missing values
for col in cat_cols:
    dataset[col] = dataset[col].fillna(dataset[col].mode()[0])
```

```python

# =========================
# Date feature engineering
# =========================
dataset['Year'] = dataset['Order Date'].dt.year
dataset['Month'] = dataset['Order Date'].dt.month
dataset['Day'] = dataset['Order Date'].dt.day
```

```python
# Drop original date column
dataset = dataset.drop(columns=['Order Date'])
    
```
